```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
#load the dataset
data=pd.read_csv('Churn_Modelling (1).csv')
data.head()
```

```python
data.info()
```

```python
# Preprocess data
# drop irrelevant features or columns
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1)
data
```

```python
# Encode categorical values
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
data.isnull().sum()
```

```python
data['Geography'].unique()
```

```python
data['Geography'].value_counts()
```

```python
data['HasCrCard'].value_counts()
```

```python
data['IsActiveMember'].value_counts()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['HasCrCard'] = data['HasCrCard'].fillna(1.0)
data['IsActiveMember'] = data['IsActiveMember'].fillna(1.0)
data['Geography'] = data['Geography'].fillna('France')
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
# One hot encoder geography column
from sklearn.preprocessing import OneHotEncoder
onehot_encoder_geo = OneHotEncoder()
geo_encoder = onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
a = geo_encoder.toarray()
```

```python
b = onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df = pd.DataFrame(a,columns=b)
geo_encoded_df
```

```python
# combine one hot encoder columns with the original data
data = pd.concat([data.drop('Geography',axis=1),geo_encoded_df],axis=1)
data.head()
```

```python
# Save the encoders and scaler
with open('label_encoder_gender.pk1','wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('onehot_encoder_geo.pk1','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
# Divide the dataset int independent(i/p) and dependent(o/p) features:
x = data.drop('Exited', axis=1)
y = data['Exited']

# split the data in training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

# scale these features:
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
with open('scaler.pk1','wb') as file:
    pickle.dump(scaler, file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import datetime
```

```python
# Build our ANN model
model = Sequential([
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test,y_test), epochs=100, callbacks=[early_stopping_callback])
```

```python
# save the model file
model.save('model.h5')
```
